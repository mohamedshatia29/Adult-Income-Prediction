```python
questionable_cat_cols = [ "workclass", "education", "marital_status", "occupation" , "relationship" , "race" , "sex" , "native_country" , "income"]

for col in questionable_cat_cols:
    print(col) 
    print(df[col].unique().tolist())
    print("-------------------------")
    print()
```

```python
questionable_num_cols = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
for col in questionable_num_cols: 
    if df[col].isna().any():
        print(col, ": have missing")
```

- I Think This dataset often includes extra spaces (e.g., " Private" → "Private") so We will strip leading and trailing spaces from all string values

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```
